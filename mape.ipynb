# Pearson R example used to evaluate Corelation between data

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

%matplotlib inline
```

### Create fake data

```python
# set random seed for repeatability
np.random.seed(1)

n_samples = 1000
x = np.linspace(0, 10, n_samples)

y = 1 * x + 10
y = np.sin(x)
mu = 0
sigma = 1
noise = np.random.normal(mu, sigma, x.shape)
y_noise = y + noise


# ploy x versus y
plt.scatter(x, y_noise)
plt.plot(x, y, color='r')
```

```python
count, bins, ignored = plt.hist(noise, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
```

```python
r, p = pearsonr(y, y_noise)
mae = np.mean(np.abs(y - y_noise))

# MAPE is bad since it's undefine if y is zero, suggest using Symetric Mean Absolute Percentage Error (SMAPE)
# https://en.wikipedia.org/wiki/Symmetric_mean_absolute_percentage_error
mape = np.mean(np.abs( ( y - y_noise) / (np.abs(y) + np.abs(y_noise))/2))

print(f"r={r}, mae={mae}, mape={mape}")
```
